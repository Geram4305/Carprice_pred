# Price vs Categorical variables

```python
df_ = pd.DataFrame(df.groupby(['enginetype'])['price'].mean().sort_values(ascending = False))
df_.plot.bar(figsize=(8,6))
plt.title('Engine Type vs Average Price')
plt.show()

df['enginetype'].hist()
```

dohcv is the most priced type
Engine type seems to be most favored type.
ohcv has the highest price range (While dohcv has only one row), ohc and ohcf have the low price range.

From histogram, it looks like ohc is the most common engine type

```python
price_brand = df.groupby(['brand'])['price'].mean().sort_values(ascending = False)
price_brand.plot(kind = 'bar')
plt.show()

fuel_price= df.groupby(['fueltype'])['price'].mean().sort_values(ascending = False)
fuel_price.plot(kind = 'bar')
plt.show()

carbody_price = df.groupby(['carbody'])['price'].mean().sort_values(ascending = False)
carbody_price.plot(kind = 'bar')
plt.show()
```

From the above plot, it looks like Jaguar and Buick are the most priced car brand in the dataset with higher avg prices compared to rest. On the other hand, Chevrolet is least priced. Similarly the average price of diesel cars exceed the gas type. Makes sense as diesel cars are expensive compared to the gas. The average price of hard top and convertile carbody types are higher.

# Price vs Numerical variables

```python
def scatter(x,fig):
    plt.subplot(2,2,fig)
    plt.scatter(df[x],df['price'])
    plt.title(x+' vs Price')
    plt.ylabel('Price')
    plt.xlabel(x)
plt.figure(figsize=(10,20))
```

```python
scatter('carlength', 1)
scatter('carwidth', 2)
scatter('carheight', 3)
scatter('curbweight', 4)

plt.tight_layout()
```

carlength,carwidth,carheight shows a positive correlation with price, whereas the curbweight doesnt look like having a trend with price

```python
def pairplot(x,y,z):
    sns.pairplot(df,x_vars = [x,y,z],y_vars='price',size=4, aspect =1, kind = 'scatter')
    plt.show()
```

```python
pairplot('enginesize', 'boreratio', 'stroke')
pairplot('compressionratio', 'horsepower', 'peakrpm')
pairplot('wheelbase', 'citympg', 'highwaympg')
```

enginesize, boreratio, horsepower, wheelbase looks like having positive correlation with price.
citympg, highwaympg on the other hand looks like having negative correlation with price.


```python
df.drop(['car_ID'],axis=1,inplace = True) 
```

```python
%matplotlib inline
plt.figure(figsize=(12,10))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
corr_Price=cor["price"].sort_values(ascending=False)
print(corr_Price)
```

Considering the correlation heatmap above and the scatterplot analysis, enginesize seems to have the highest positive correlation with price
