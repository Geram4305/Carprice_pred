```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
np.random.seed(0) 
```

```python
df = pd.read_csv('/data/CarPrice.csv')
df.head()
```

```python
df.drop(['car_ID'],axis=1,inplace = True) 
```

```python
brand = df['CarName'].apply(lambda x : x.split(' ')[0])
```

```python
df.insert(3,"brand",brand)
```

```python
df.drop('CarName',axis = 1,inplace=True)
```

```python
df.head()
```

```python
df.select_dtypes(include=['object'])
```

```python
# Defining the map function
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
# Applying the function to the cars_lr

df = dummies('fueltype',df)
df = dummies('aspiration',df)
df = dummies('carbody',df)
df = dummies('drivewheel',df)
df = dummies('enginetype',df)
df = dummies('cylindernumber',df)
df = dummies('brand',df)
df = dummies('enginelocation',df)
df = dummies('fuelsystem',df)
```

```python
df.shape
```

```python
df.columns.values
```

```python
y= df.pop('price')
X= df
```

'enginesize', 'curbweight', 'highwaympg','horsepower', 'carwidth'are the top 5 important features by random forest

Hyperparameter tuning for selecting the best random forest model

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200,num = 12)]
max_features = ['auto','sqrt']
max_depth = [int(x) for x in np.linspace(5,30,num = 6)]
min_samples_split = [2,5,10,15,100]
min_samples_leaf  = [1,2,5,10]
grid = {'n_estimators': n_estimators,
        'max_features': max_features,
        'max_depth': max_depth,
        'min_samples_split': min_samples_split,
       'min_samples_leaf': min_samples_leaf}
print(grid)
```

```python
x_train, x_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
model = RandomForestRegressor() 
hyp = RandomizedSearchCV(estimator = model,
                        param_distributions=grid,
                        n_iter=10,
                       scoring='neg_mean_squared_error',
                        cv=5,verbose = 2,
                        random_state = 42,n_jobs = 1) 
hyp.fit(x_train,y_train)
```

```python
print('Best Score: %s' % hyp.best_score_)
print('Best Hyperparameters: %s' % hyp.best_params_)
```

Best Score is -5408089.372248346 with parameters 'n_estimators': 700, 'min_samples_split': 15, 'min_samples_leaf': 1, 'max_features': 'auto', 'max_depth': 20

```python
y_pred = hyp.predict(x_test)
```
